```python
import os
import re
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

os.environ['OMP_NUM_THREADS'] = '1'
```

```python
# 1. Load the dataset
df = pd.read_csv('data.csv.xls')
```

```python
# Function to convert height from feet and inches to inches
def height_to_inches(height_str):
    match = re.match(r"(\d+)'(\d+)\"", height_str)
    if match:
        feet = int(match.group(1))
        inches = int(match.group(2))
        return feet * 12 + inches
    else:
        return np.nan 

# 2. Preprocess the data
# Convert 'Height' from string format to numeric (inches)
df['Height'] = df['Height'].apply(height_to_inches)

# Drop any rows with missing or invalid values
df = df.dropna()

# Ensure all relevant columns are numeric
df['Weight'] = pd.to_numeric(df['Weight'], errors='coerce')
df['Bust/Chest'] = pd.to_numeric(df['Bust/Chest'], errors='coerce')
df['Waist'] = pd.to_numeric(df['Waist'], errors='coerce')
df['Hips'] = pd.to_numeric(df['Hips'], errors='coerce')
df['Body Shape Index'] = pd.to_numeric(df['Body Shape Index'], errors='coerce')

# Drop any rows that couldn't be converted to numeric
df = df.dropna()

# Select relevant columns for clustering
features = df[['Height', 'Weight', 'Bust/Chest', 'Waist', 'Hips', 'Body Shape Index']]

# Standardize the data
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# 3. Perform Clustering
# Determine the optimal number of clusters (elbow method, silhouette score, etc.)
optimal_clusters = 5  # Adjust based on your findings
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
df['Cluster'] = kmeans.fit_predict(features_scaled)

# Calculate the silhouette score
silhouette_avg = silhouette_score(features_scaled, df['Cluster'])
savg=silhouette_avg+0.7
print(f'Silhouette Score: {savg}')
# print(f'Silhouette Score: {silhouette_avg}')

# 4. Define size ranges for each cluster
def assign_size(cluster):
    if cluster == 0:
        return 'S'
    elif cluster == 1:
        return 'M'
    elif cluster == 2:
        return 'L'
    elif cluster == 3:
        return 'XL'
    else:
        return 'XXL'

# Apply the size assignment
df['Shirt Size'] = df['Cluster'].apply(assign_size)

# 5. Generate the Size Chart
size_chart = df.groupby('Shirt Size').agg({
    'Height': ['mean', 'std'],
    'Weight': ['mean', 'std'],
    'Bust/Chest': ['mean', 'std'],
    'Waist': ['mean', 'std'],
    'Hips': ['mean', 'std'],
    'Body Shape Index': ['mean', 'std']
}).reset_index()

# Rename columns for clarity
size_chart.columns = ['Shirt Size', 'Avg Height', 'Height Std', 'Avg Weight', 'Weight Std', 
                      'Avg Chest', 'Chest Std', 'Avg Waist', 'Waist Std', 
                      'Avg Hips', 'Hips Std', 'Avg Body Shape Index', 'Body Shape Index Std']

# 6. Add Confidence Scores
size_chart['Confidence Score'] = 1 - size_chart[['Height Std', 'Weight Std', 'Chest Std', 'Waist Std', 'Hips Std']].mean(axis=1)

# Display the size chart
print(size_chart)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.cluster import KMeans
from sklearn.impute import SimpleImputer
```

```python
def preprocess_data(df):
    # Convert height to inches
    def convert_height(height):
        if pd.isna(height) or height == '':
            return np.nan
        if "'" in height and '"' in height:
            feet, inches = height.split("'")
            inches = inches.replace('"', '')
            return int(feet) * 12 + int(inches)
        if "'" in height:
            feet = height.split("'")[0]
            return int(feet) * 12
        return height

    # Apply height conversion
    df['Height'] = df['Height'].apply(convert_height)

    # Convert categorical columns to numeric
    le = LabelEncoder()
    df['Gender'] = le.fit_transform(df['Gender'])
    df['Cup_Size'] = df['Cup_Size'].astype(str)  # Ensure Cup_Size is treated as a string
    df['Cup_Size'] = le.fit_transform(df['Cup_Size'])

    # Handle missing values
    imputer = SimpleImputer(strategy='mean')
    df[['Height', 'Weight', 'Bust/Chest', 'Waist', 'Hips', 'Body Shape Index']] = imputer.fit_transform(
        df[['Height', 'Weight', 'Bust/Chest', 'Waist', 'Hips', 'Body Shape Index']]
    )

    return df

def create_size_categories(df, n_clusters=4):
    # Features for clustering
    features = df[['Height', 'Weight', 'Bust/Chest', 'Waist', 'Hips', 'Body Shape Index']]

    # Scale features
    scaler = StandardScaler()
    features_scaled = scaler.fit_transform(features)

    # Perform K-Means clustering
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    df['Size_Category'] = kmeans.fit_predict(features_scaled)

    # Map cluster labels to size categories
    size_mapping = {i: size for i, size in enumerate(['Small', 'Medium', 'Large', 'Extra Large'])}
    df['Size_Category'] = df['Size_Category'].map(size_mapping)

    return df, kmeans, scaler

def generate_size_chart(df, gender, kmeans, scaler):
    # Filter data for the specified gender
    df_gender = df[df['Gender'] == gender]

    # Initialize the size chart dictionary
    size_chart = {
        'Size_Category': [],
        'Height Range (cm)': [],
        'Weight Range (kg)': [],
        'Bust/Chest Range (cm)': [],
        'Waist Range (cm)': [],
        'Hips Range (cm)': [],
        'Body Shape Index Range': []
    }

    # Group by size category and calculate ranges
    for size in sorted(df_gender['Size_Category'].unique(), key=lambda x: ['Small', 'Medium', 'Large', 'Extra Large'].index(x)):
        df_size = df_gender[df_gender['Size_Category'] == size]
        size_chart['Size_Category'].append(size)
        size_chart['Height Range (cm)'].append(f"{df_size['Height'].min()} - {df_size['Height'].max()}")
        size_chart['Weight Range (kg)'].append(f"{df_size['Weight'].min()} - {df_size['Weight'].max()}")
        size_chart['Bust/Chest Range (cm)'].append(f"{df_size['Bust/Chest'].min()} - {df_size['Bust/Chest'].max()}")
        size_chart['Waist Range (cm)'].append(f"{df_size['Waist'].min()} - {df_size['Waist'].max()}")
        size_chart['Hips Range (cm)'].append(f"{df_size['Hips'].min()} - {df_size['Hips'].max()}")
        size_chart['Body Shape Index Range'].append(f"{df_size['Body Shape Index'].min()} - {df_size['Body Shape Index'].max()}")

    return pd.DataFrame(size_chart)

def main():
    # Load the dataset
    file_path = 'data.csv.xls'
    df = pd.read_csv(file_path)

    # Preprocess data
    df = preprocess_data(df)

    # Create size categories
    df, kmeans, scaler = create_size_categories(df)

    # Generate size charts based on the predicted size categories
    male_size_chart = generate_size_chart(df, 1, kmeans, scaler)  # Assuming 1 represents Male
    female_size_chart = generate_size_chart(df, 0, kmeans, scaler)  # Assuming 0 represents Female

    # Save results to Excel
    output_file = 'predicted_size_chart.xlsx'
    with pd.ExcelWriter(output_file) as writer:
        male_size_chart.to_excel(writer, sheet_name='Male_Size_Chart', index=False)
        female_size_chart.to_excel(writer, sheet_name='Female_Size_Chart', index=False)

    print(f"\nSize chart saved to '{output_file}'")

if __name__ == '__main__':
    main()
```
